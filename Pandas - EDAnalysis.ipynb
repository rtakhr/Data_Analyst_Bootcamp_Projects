```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#format to two decimal places
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df.describe()
```

```python
#check null values
df.isnull().sum()
```

```python
df.nunique()
```

```python
#view top range of countries
df.sort_values(by = "World Population Percentage", ascending = False).head(10)
```

```python
#viewing correlations in numeric values
df.corr(numeric_only = True)
```

```python
#visualize correlation using heatmap
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams['figure.figsize'] = (20,6)

plt.show()
```

```python
#Grouping data (continents)
df.groupby("Continent").mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
#Visualizing data above
#use only population over time and in ascending order
df2 = df.groupby("Continent")[df.columns[5:13][::-1]].mean(numeric_only = True).sort_values(by = "2022 Population", 
                                                                                            ascending = False)
df2
```

```python
#switch rows and columns for better visualization
df3 = df2.transpose()
```

```python
df3.plot()
```

```python
#boxplotting outliers
df.boxplot(figsize = (20,6))
```
