# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\takhz\Desktop\Pandas\Customer Call List.xlsx")
df
```

```python
#remove duplicates in data set
df = df.drop_duplicates()
```

```python
#remove unnecessary columns (the last column)
df = df.drop(columns = 'Not_Useful_Column')
```

```python
#clean last_name column
#df["Last_Name"] = df["Last_Name"].str.strip("...")
#df["Last_Name"] = df["Last_Name"].str.strip("/")
#df["Last_Name"] = df["Last_Name"].str.strip("_")

#or
df["Last_Name"] = df["Last_Name"].str.strip("_/...")
df
```

```python
#clean phone_number column (using th first item's format)
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
#clean address column (split the column)
df[['Street_Address', 'State', 'Zip_Code']] = df["Address"].str.split(',', 2, expand = True)
#we can delete the address column after this
df = df.drop(columns = 'Address')
df
```

```python
#clean paying customer and do not contact columns
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
#clean null values
df = df.fillna('')
df = df.replace('N/a','')
df
```

```python
#Additional Task: Provide a list of phone numbers they can call (remove rows with Y in do not call and rows with no numbers)
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x,  inplace = True)
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x,  inplace = True)
df

#or
#df.dropna(subset = "Phone_Number", inplace = True)
```

```python
#reset index
df = df.reset_index(drop = True)
df
```
